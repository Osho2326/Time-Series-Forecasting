### **ARIMA Model**

```python
!pip install pmdarima
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_log = np.log(df_tesla['Close'])
moving_avg = df_log.rolling(12).mean()
std_dev = df_log.rolling(12).std()
plt.legend(loc='best')
plt.title('Moving Average')
plt.plot(std_dev, color ="black", label = "Standard Deviation")
plt.plot(moving_avg, color="red", label = "Mean")
plt.show()
```

```python
train_data, test_data = df_log[3:int(len(df_log)*0.9)], df_log[int(len(df_log)*0.9):]
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('Dates')
plt.ylabel('Closing Prices')
plt.plot(df_log, 'green', label='Train data')
plt.plot(test_data, 'red', label='Test data')
plt.legend()
```

```python
from pmdarima.arima import auto_arima
model_autoARIMA = auto_arima(train_data, start_p=0, start_q=0,
test='adf',       # use adftest to find optimal 'd'
max_p=3, max_q=3, # maximum p and q
m=1,              # frequency of series
d=None,           # let model determine 'd'
seasonal=False,   # No Seasonality
start_P=0,
D=0,
trace=True,
error_action='ignore',
suppress_warnings=True,
stepwise=True)
print(model_autoARIMA.summary())
```

```python
model_autoARIMA.plot_diagnostics(figsize=(15,8))
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(train_data, order=(3, 1, 2))
fitted = model.fit()
print(fitted.summary())
```

```python
# Forecasting 282 steps ahead
forecast_obj = fitted.get_forecast(steps=282)

# Get forecasted values
fc = forecast_obj.predicted_mean

# Get confidence intervals
conf = forecast_obj.conf_int(alpha=0.05)

# Get standard errors
se = forecast_obj.se_mean
```

```python
# Forecast object should have `forecast` method to get intervals
forecast_obj = fitted.get_forecast(steps=len(test_data))
fc = forecast_obj.predicted_mean
conf = forecast_obj.conf_int(alpha=0.05)
```

```python
# Align the forecast series with test_data index
fc_series = pd.Series(fc, index=test_data.index[:len(fc)])
lower_series = pd.Series(conf.iloc[:, 0].values, index=test_data.index[:len(fc)])
upper_series = pd.Series(conf.iloc[:, 1].values, index=test_data.index[:len(fc)])

# Plotting
plt.figure(figsize=(12, 5), dpi=100)
plt.plot(train_data, label='Training Data')
plt.plot(test_data, color='blue', label='Actual Stock Price')
plt.plot(fc_series, color='orange', label='Predicted Stock Price')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.10)
plt.title('Tesla Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend(loc='upper center', fontsize=8)
plt.show()
```
